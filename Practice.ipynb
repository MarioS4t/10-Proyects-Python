```python
from geopy.geocoders import ArcGIS
nom = ArcGIS()
```

```python
nom.geocode("3995 23rd St, San Francisco, CA 94114")
```

```python
df = pd.read_csv('supermarkets/supermarkets.csv')
df
```

```python
df["Address"] = df["Address"]+ ", " + df["City"] + ", " + df["State"] + ", " + df["Country"]
```

```python
df
```

```python
df["Coordedinates"]= df["Address"].apply(nom.geocode)
df.Coordedinates[0].latitude
```

```python
df["Latitude"] = df["Coordedinates"].apply(lambda x: x.latitude if x != None else None)
df["Longitude"] = df["Coordedinates"].apply(lambda x: x.longitude if x != None else None)
```
